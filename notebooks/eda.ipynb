```python
import pandas as pd
import numpy as np

csv_path = pd.read_csv('../data/cleaned_data.csv')
csv_path.head()
```

```python
csv_path['name'] = csv_path['name'].str.strip().str.title()
csv_path = csv_path.rename(columns={'name': 'product_name'})

```

```python
csv_path.dtypes
```

```python
csv_path['revenue'].isna().sum()
```

```python
csv_path["products_total_revenue"] = csv_path.groupby('product_name')['revenue'].transform('sum')
```

```python
csv_path["supplier_total_revenue"] = csv_path.groupby('supplier')['revenue'].transform('sum')
```

```python
csv_path["category_total_revenue"] = csv_path.groupby('category')['revenue'].transform('sum')
csv_path.head()
```

```python
revenue_generated = csv_path[["products_total_revenue", "supplier_total_revenue", "category_total_revenue"]]
revenue_generated
```

```python

csv_path['timestamp'] = pd.to_datetime(csv_path['timestamp'])
csv_path['month'] = csv_path['timestamp'].dt.to_period('M')

monthly_trend = csv_path.pivot_table(
    values='revenue',
    index='month',
    aggfunc='sum'
)

```
